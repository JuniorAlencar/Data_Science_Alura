```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('properties.mplstyle')
```

```python
df = pd.read_csv("../canadian_immegration_data.csv")
df.set_index('Country', inplace=True)
anos = list(map(str, range(1980, 2014)))

brasil = df.loc["Brazil", anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)

argentina = df.loc["Argentina", anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
```

```python

thickness_axes = 1.4
font_legend = 12

# Modificando a espessura da box do plot
ax = plt.gca()
parts_of_graphic = ['top', 'bottom', 'left', 'right']
for k in parts_of_graphic:
    ax.spines[k].set_linewidth(thickness_axes)  # substitua 2 por thickness_axes

# Filtrando o dataframe, para encontrar em qual ano houve o maior nmero de imigrantes
max_val = dados_brasil["imigrantes"].max()
max_imigrantes = [(ano, imigra) for ano, imigra in zip(dados_brasil["ano"], dados_brasil["imigrantes"]) if imigra == max_val]

plt.plot(dados_brasil["ano"], dados_brasil["imigrantes"],'-o', lw=2*thickness_axes, mew=2*thickness_axes)
plt.plot(max_imigrantes[0][0], max_imigrantes[0][1],'o')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.xlabel("ano")
plt.ylabel("Nmero de Imigrantes")
plt.title("Imigrao do Brasil para o Canad")
plt.show()
```
