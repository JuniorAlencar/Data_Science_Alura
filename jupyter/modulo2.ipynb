```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('properties.mplstyle')
```

```python
df = pd.read_csv("../canadian_immegration_data.csv")
df.set_index('Country', inplace=True)
anos = list(map(str, range(1980, 2014)))

brasil = df.loc["Brazil", anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)

argentina = df.loc["Argentina", anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
```

```python
# Lojas
lojas = ['A', 'B', 'C', 'D']

# DicionÃ¡rio com os dados
vendas_2022 = {
    'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# DicionÃ¡rio para armazenar os dados
dict_vendas = {"loja": [], "mes": [], "vendas": []}

# Iterar sobre o dicionÃ¡rio
for mes, vals in vendas_2022.items():
    for loja, venda in zip(lojas, vals):
        dict_vendas["mes"].append(mes)
        dict_vendas["loja"].append(loja)
        dict_vendas["vendas"].append(venda)

# Construir DataFrame
df_vendas = pd.DataFrame(data=dict_vendas)

# Definir ordem dos meses
ordem_meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
               'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']

# Ordenando para apresentar de Janeiro atÃ© Dezembro em ordem cronologica
df_vendas["mes"] = pd.Categorical(df_vendas["mes"],
                                  categories=ordem_meses,
                                  ordered=True)

# Ordenar por loja e depois por mÃªs
df_vendas = df_vendas.sort_values(by=["loja", "mes"]).reset_index(drop=True)
df_vendas
```

```python
# ordem correta dos meses
ordem_meses = ['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun',
               'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez']
# Importante salientar o funcionamento do Categorical define uma ordem, seguindo a estrutura de entrada (ordem meses)
# Com isso, ele declara Jan > Fev > Mar, etc.
df_vendas['mes'] = pd.Categorical(df_vendas['mes'], categories=ordem_meses, ordered=True)

# Tabela com linhas=meses e colunas=lojas
tab = df_vendas.pivot(index='mes', columns='loja', values='vendas').loc[ordem_meses]

# eixo X numÃ©rico
x = np.arange(len(ordem_meses))

fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Vendas por loja e mÃªs em 2022')

lojas = ['A','B','C','D']
thickness_axes = 1.4
font_ticks = 14
# Cria um zip onde ax faz paret do axs.flat, onde o .flat, torna ax[0,0], ax[0,1], ax[1,0] e ax[1,1] numa lista flat ax[0,0] ->1, ax[0,1] ->2, etc
for ax, loja in zip(axs.flat, lojas):
    y = tab[loja].to_numpy()             # garante array numÃ©rico
    ax.plot(x, y, marker='o')
    ax.set_title(loja)

    # meses como labels
    ax.set_xticks(x)
    ax.set_xticklabels(ordem_meses, rotation=0)

    ax.tick_params(which='minor', width=thickness_axes, length=4, direction="in", labelsize=font_ticks)
    ax.tick_params(which='major', width=thickness_axes, length=8, direction="in", labelsize=font_ticks)
    ax.set_ylim([0, 450])
plt.show()
```
