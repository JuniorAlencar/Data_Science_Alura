### Irei fazer uma animao, onde o nmero de imigrantes total ser atualizado a cada ano, de 1980 at 2013. Para que Possamos acompanhar visualmente como se deu esse processo imigratrio ao longo do tempo. Para isso, irei pegar o dataframe da America do sul, somando o nmero de imigrantes a cada ano e atualizando o plot

```python
df = pd.read_csv("../canadian_immegration_data.csv")

# Listagem dos anos
anos = list(map(str, range(1980, 2014)))

# Filtrando os pases da America do Sul
df = df.query('Region == "South America"').reset_index(drop=True)

# Anos de interesse
anos = [str(y) for y in range(1980, 2014)]  # 1980..2013

# Pega s Country + anos, faz cumulativo por linha (pas), e volta o ndice
df_cumulativo = (
    df[["Country"] + anos]
    .set_index("Country")[anos]
    .apply(pd.to_numeric, errors="coerce")  # garante numrico
    .cumsum(axis=1)                         # cumulativo ao longo dos anos
    .reset_index()
)
# Com isso temos o dataframe com o nmero de imigrantes total ao longo de cada ano
df_cumulativo
```

```python
# df_cumulativo: colunas ['Country','1980',...,'2013']
anos = [str(y) for y in range(1980, 2014)]
countries = df_cumulativo["Country"].tolist()
n = len(countries)

# Cores fixas por pas
palette = ['royalblue','orange','forestgreen','orchid','purple','brown',
           'slateblue','gray','olive','navy','teal','tomato']
color_map = {c: palette[i % len(palette)] for i, c in enumerate(countries)}

# Matriz (n_paises x n_anos)
vals = df_cumulativo.set_index("Country")[anos].loc[countries].values


def ranks_for_year(values, ascending=True):
    order = np.argsort(values) if ascending else np.argsort(-values)
    rank = np.empty_like(order)
    rank[order] = np.arange(len(values))
    return rank

# ---- Config ----
steps = 24          # MAIS passos = transio mais lenta/suave
fps = 15            # s usado ao salvar; no afeta plt.show
ascending = True

# ---- Figure ----
fig, ax = plt.subplots(figsize=(19, 8), constrained_layout=True)

ax.set_title(
    "Imigrao da Amrica do Sul para o Canad (19802013)",
    loc='left',
    fontsize=25
)

# MAIS ESPAO  ESQUERDA p/ no cortar nomes
fig.subplots_adjust(left=0.25)   # aumente para 0.45 se ainda cortar

ax.set_xlabel('Nmero de imigrantes', fontsize=25)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=22)
ax.yaxis.set_tick_params(labelsize=22)

yticks = np.arange(n)
ax.set_yticks(yticks)
ax.set_yticklabels(countries)

# barras iniciais (1980)
k0 = 0
v0 = vals[:, k0]
r0 = ranks_for_year(v0, ascending=ascending).astype(float)

bars = ax.barh(yticks, v0, height=0.8, color=[color_map[c] for c in countries])
for bar, y in zip(bars, r0):
    h = bar.get_height()          # 0.8
    bar.set_y(y - h/2)            # centraliza no tick


# Texto do ano no canto inferior direito (SEM %)
year_txt = ax.text(0.98, 0.02, f"Ano = {anos[k0]}", transform=ax.transAxes,
                   ha='right', va='bottom', fontsize=30)

def set_xlim_padding(v):
    vmax = float(np.nanmax(v)) if np.all(np.isfinite(v)) else 1.0
    ax.set_xlim(0, vmax * 1.1 if vmax > 0 else 1.0)

set_xlim_padding(v0)

# Funo para atualizar os frames da animao
def update(frame):
    k = frame // steps
    t = (frame % steps) / steps

    if k >= len(anos) - 1:
        vk = vals[:, -1]
        rk = ranks_for_year(vk, ascending=ascending).astype(float)
        order = np.argsort(vk) if ascending else np.argsort(-vk)

        for bar, y, w in zip(bars, rk, vk):
            bar.set_width(w)
            bar.set_y(y)

        # Atualiza labels para alinhar com a ordem final
        ax.set_yticklabels([countries[i] for i in order])

        year_txt.set_text(f"Ano = {anos[-1]}")
        set_xlim_padding(vk)
        return [*bars, year_txt]

    vA, vB = vals[:, k], vals[:, k + 1]
    rA = ranks_for_year(vA, ascending=ascending).astype(float)
    rB = ranks_for_year(vB, ascending=ascending).astype(float)

    v = (1 - t) * vA + t * vB
    r = (1 - t) * rA + t * rB

    for bar, y, w in zip(bars, r, v):
        bar.set_width(w)
        h = bar.get_height()
        bar.set_y(y - h/2)            # centraliza


    # Obtm ordem atual para labels
    order = np.argsort(v) if ascending else np.argsort(-v)
    ax.set_yticklabels([countries[i] for i in order])

    year_txt.set_text(f"Ano = {anos[k]}")
    set_xlim_padding(v)

    return [*bars, year_txt]


total_frames = (len(anos) - 1) * steps + 1

# INTERVALO MAIOR = mais lento (ms por frame)
anim = FuncAnimation(fig, update, frames=total_frames, interval=120, blit=True)

plt.show()

# Para salvar devagarzinho:
anim.save('../resultados/imigracao_bar_race.mp4', fps=fps, dpi=150)
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(8, 4))

lojas = ['A', 'B', 'C', 'D']

vendas_2022 = {'Jan': [100, 80, 150, 50],
    'Fev': [120, 90, 170, 60],
    'Mar': [150, 100, 200, 80],
    'Abr': [180, 110, 230, 90],
    'Mai': [220, 190, 350, 200],
    'Jun': [230, 150, 280, 120],
    'Jul': [250, 170, 300, 140],
    'Ago': [260, 180, 310, 150],
    'Set': [240, 160, 290, 130],
    'Out': [220, 140, 270, 110],
    'Nov': [400, 220, 350, 190],
    'Dez': [300, 350, 400, 250]
}

# Criando DataFrame
df = pd.DataFrame(vendas_2022, index=lojas)

# Interando sobre as linhas, onde o subset representa o sub-conjunto do dataframe (dados da loja especifica)
for loja, subset in df.iterrows():
    # Com esse mtodo, os novos indices sero os meses, com os valores sendo atribuido aos values
    meses = subset.index        # conjunto de meses
    valores = subset.values     # conjunto de valores
    ax.plot(meses, valores, 'o-', label=f'Loja {loja}')

ax.set_title('Vendas para o ano de 2022 para as quatro lojas', fontsize=20, loc='left')
ax.set_ylabel('Nmero de vendas', fontsize=14)
ax.set_xlabel('Ms', fontsize=14)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_tick_params(labelsize=12)
plt.legend()
plt.show()
```

Podemos observar que as quatro lojas possuem um padro de crescimento muito similar, onde Janeiro comea com um baixo nmero de vendas, que aumenta gradativamente at Abril, com um pico subito no ms de maio, decaindo em Junho. Outubro h um mnimo com um crescimento considervel em Novembro e Dezembro.
