```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
import matplotlib as mpl
plt.style.use('properties.mplstyle')
```

Os dados se referem as imigrações para o Canadá, tendo o país de saída "Country".

```python
df = pd.read_csv("../canadian_immegration_data.csv")
df
```

```python
df.info()
```

```python
df.set_index('Country', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
brasil = df.loc["Brazil", anos]
brasil_dict = {'ano': brasil.index.tolist(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python

thickness_axes = 1.4
font_legend = 12

# Modificando a espessura da box do plot
ax = plt.gca()
parts_of_graphic = ['top', 'bottom', 'left', 'right']
for k in parts_of_graphic:
    ax.spines[k].set_linewidth(thickness_axes)  # substitua 2 por thickness_axes

# Filtrando o dataframe, para encontrar em qual ano houve o maior número de imigrantes
max_val = dados_brasil["imigrantes"].max()
max_imigrantes = [(ano, imigra) for ano, imigra in zip(dados_brasil["ano"], dados_brasil["imigrantes"]) if imigra == max_val]

plt.plot(dados_brasil["ano"], dados_brasil["imigrantes"])
plt.plot(max_imigrantes[0][0], max_imigrantes[0][1],'o')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.xlabel("ano")
plt.ylabel("Número de Imigrantes")
plt.title("Imigração do Brasil para o Canadá")
plt.show()
```

```python
anos = list(map(str, range(1980, 2014)))
argentina = df.loc["Argentina", anos]
argentina_dict = {'ano': argentina.index.tolist(), 'imigrantes': argentina.values.tolist()}
dados_argentina = pd.DataFrame(argentina_dict)
dados_argentina
```

```python
thickness_axes = 1.4
font_legend = 12
font_ticks = 13
font_eixos = 13

# Modificando a espessura da box do plot
ax = plt.gca()
parts_of_graphic = ['top', 'bottom', 'left', 'right']
for k in parts_of_graphic:
    ax.spines[k].set_linewidth(thickness_axes)  # substitua 2 por thickness_axes

# Plot dos dados da Argetina e do Brasil
plt.plot(dados_brasil["ano"], dados_brasil["imigrantes"], label='Brasil')
plt.plot(dados_argentina["ano"], dados_argentina["imigrantes"], label="Argentina")

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

# Labels dos eixos
plt.xlabel("ano")
plt.ylabel("Número de Imigrantes")
plt.title("Imigração do Brasil e Argentina para o Canadá")
# Propriedades da legenda
plt.legend(prop = {"size":font_legend},framealpha=0.0)

plt.tick_params(which='minor', width=thickness_axes, length=4,direction="in",labelsize=font_ticks)
plt.tick_params(which='major', width=thickness_axes ,length=8,direction="in",labelsize=font_ticks)

plt.show()
```

Entre os anos de 1985 e 1990 houve um crescimento consideravelmente grande de imigração, tanto na Argetina, quanto no Brasil. Com um decrescimento enter 1990 e 1995. Tanto no Brasil quanto na Argetina, houve uma estagnação entre 1995 e 2000. Na Argetina, entre 2000 e 2005, houve um crescimento consideravelmente grande na Argetina, atingindo um máximo, enquanto no Brasil, esse crescimento atingindo um máximo ocorreu entre 2005 e 2010.

Irei criar um novo dataframe, partindo dos dados do Brasil e da Argetina. Nesse novo dataframe, ele irá carregar as diferenças de imigrações entre o Brasil e a Argetina para o Canadá. Vale ressaltar, que o sinal negativo, representará uma maior imigração da Argetina, em contrapartida o sinal positivo uma maior imigração do Brasil.

```python
# Plot dos dados da Argetina e do Brasil
dif_data = dados_brasil["imigrantes"] - dados_argentina["imigrantes"]
anos = list(map(str, range(1980, 2014)))
dif_dict = {'ano': anos, 'imigrantes': dif_data}
dados_diferenca = pd.DataFrame(dif_dict)
dados_diferenca
```

```python
thickness_axes = 1.4

# Modificando a espessura da box do plot
ax = plt.gca()
parts_of_graphic = ['top', 'bottom', 'left', 'right']
for k in parts_of_graphic:
    ax.spines[k].set_linewidth(thickness_axes)  # substitua 2 por thickness_axes

# Filtrando o dataframe, para encontrar em qual ano houve o maior número de imigrantes
max_val = dados_diferenca["imigrantes"].max()
max_imigrantes = [(ano, imigra) for ano, imigra in zip(dados_diferenca["ano"], dados_diferenca["imigrantes"]) if imigra == max_val]

# Filtrando o dataframe, para encontrar em qual ano houve o menor número de imigrantes
min_val = dados_diferenca["imigrantes"].min()
min_imigrantes = [(ano, imigra) for ano, imigra in zip(dados_diferenca["ano"], dados_diferenca["imigrantes"]) if imigra == min_val]

plt.plot(dados_diferenca["ano"], dados_diferenca["imigrantes"],label="Imigrantes(BR-AR)")
plt.plot(max_imigrantes[0][0],max_imigrantes[0][1],"o",label=f"ano: {max_imigrantes[0][0]}, $\\Delta$: {max_imigrantes[0][1]}")
plt.plot(min_imigrantes[0][0],min_imigrantes[0][1],"o",label=f"ano: {min_imigrantes[0][0]}, $\\Delta$: {abs(min_imigrantes[0][1])}")
plt.hlines(y=0,xmin=min(dados_brasil["ano"]),xmax=max(dados_brasil["ano"]), linestyles='--', color='k')

plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])

# Labels dos eixos
plt.xlabel("ano")
plt.ylabel(r"$\Delta$ imigrantes (Brasil - Argentina)")
plt.title("Diferença do número de Imigrantes entre Brasil e Argentina")
# Propriedades da legenda
plt.legend(framealpha=0.0)
plt.show()
```

Percebemos que no ano de 2003, houve a maior diferença no número de imigrantes entre o Brasil e a Argetina, contudo, desse ano em diante, a tendência foi que o número de imigração na Argetina foi decaindo, em comparação as imigrações brasileira. Até atingir um pico em 2010, com uma diferença de 2139 imigrações. Esse grande aumento das imigrações da Argentina, possivelmente foram movidas devidos a uma forte crise financeira que ocorreu entre 2001 e 2002, além do aumento consideravel da pobreza no país. Somado a isso, em 2002 entrou em vigor a Lei de Imigração e Progeção de Refugiados (IRPA) no Canadá.
